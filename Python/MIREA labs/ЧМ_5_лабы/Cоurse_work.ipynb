```python
import numpy as np
import math
import matplotlib.pyplot as plt
import sympy as sp
x = sp.symbols('x')
eps = 0.001
```

```python
def makeLagFunc(inpX, inpY):
    a = []
    for number in range(len(inpX)):
        sum = 1
        for i in range(len(inpX)):
            if (i != number):
                sum *= (x - inpX[i]) / (inpX[number] - inpX[i])   
        a.append(inpY[number] * sum)
    sum = 0
    for number in range(len(a)):
        sum += a[number]
    return sum
```

```python
def printLagFunc(inpX, f):
    stepMultiply = 100
    plt.grid(True)
    dist = int((inpX[-1] - inpX[0]) * stepMultiply + 1)
    X = []
    Y = []
    for i in range(dist):
        X.append(inpX[0] + i / stepMultiply)
        Y.append(f.subs(x, X[-1]))
    plt.plot(X, Y, color = 'blue')
    plt.show()
```

```python
def gradientDescent(f, xdot, inpX, eps):
    x1 = [xdot]
    a = sp.diff(f, x)
    x1.append(x1[-1] - eps * 5 * a.subs(x, x1[-1]))
    i = 0
    while (abs(f.subs(x, x1[-2]) - f.subs(x, x1[-1])) > eps or (a.subs(x, x1[-1]) / a.subs(x, x1[-2])) > 0):
        if x1[-1] - eps * 5 * a.subs(x, x1[-1]) < inpX[0]:
            x1.append(inpX[0])
        elif x1[-1] - eps * 5 * a.subs(x, x1[-1]) > inpX[-1]:
            x1.append(inpX[-1])
        else:
            x1.append(x1[-1] - eps * 5 * a.subs(x, x1[-1]))
        if i % 10 == 0:
            print("x",  int(i / 10), " = ", x1[-1], "; y", int(i / 10), " = ", f.subs(x,x1[-1]), sep = "")
        i += 1
        if abs(a.subs(x, x1[-1]) - a.subs(x, x1[-2])) < eps:
            break
    print("x =", x1[-1], "y =", f.subs(x,x1[-1]), len(x1))
```

```python
inp1X = [1, 1.5, 2, 2.5, 3]
inp1Y = [0.741, 0.638, 0.549, 0.472, 0.407]
f1 = sp.simplify(makeLagFunc(inpX=inp1X, inpY= inp1Y))
plt.title("First Lagrange function P1(x)")
print(f1)
printLagFunc(inpX=inp1X, f=f1)
```

```python
inp2X = [1, 1.5, 2, 2.5, 3]
inp2Y = [0.273, 0.042, -0.227, -0.288, 0.084]
plt.title("Second Lagrange function P2(x)")
f2 = sp.simplify(makeLagFunc(inpX=inp2X, inpY=inp2Y))
print(f2)
printLagFunc(inpX=inp2X, f=f2)
```

```python
f3 = f1 - f2
print(f3)
# print(sp.solve(f3.diff(x)))
plt.title("F(x)")
printLagFunc(inpX=inp1X, f=f3)
```

```python
gradientDescent(f=f3, xdot=2.4, inpX=inp1X, eps=eps)
```
