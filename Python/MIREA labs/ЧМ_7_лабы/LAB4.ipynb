```python
import sympy
import numpy as np
import math
import matplotlib.pyplot as plt
x = sympy.symbols('x')
y = sympy.symbols('y')
z = sympy.symbols('z')
func = (-(x**2 + 2) * y + 2*x*z)/x**2 
y1 = math.pi / 2 
z1 = 1 - math.pi / 2 
startX = math.pi / 2 
finishX = math.pi / 2 + 1 
```

```python
def graphic(a, b):
    plt.grid(True)
    plt.plot(a[0], a[1], color = 'red')
    plt.plot(b[0], b[1], color = 'blue')
    x1,x2,y1,y2 = plt.axis()
    plt.show()
```

```python
def rungeKutta(v, h, arrayX, arrayY, arrayZ):
    for i in range(v):
        print("x", i, " = ", arrayX[-1],"  y", i, " = ", arrayY[-1], sep = "")
        k1 = h*arrayZ[-1]
        l1 = h*func.subs([(x, arrayX[-1]),(y, arrayY[-1]),(z, arrayZ[-1])])
        k2 = h*(arrayZ[-1]+l1 / 2)
        l2 = h*func.subs([(x, arrayX[-1] + h / 2),(y, arrayY[-1]+ k1 / 2),(z, arrayZ[-1] + l1 / 2)])
        k3 = h*(arrayZ[-1]+l2 / 2)
        l3 = h*func.subs([(x, arrayX[-1] + h / 2),(y, arrayY[-1]+ k2 / 2),(z, arrayZ[-1] + l2 / 2)])
        k4 = h*(arrayZ[-1]+l3)
        l4 = h*func.subs([(x, arrayX[-1] + h),(y, arrayY[-1]+ k3),(z, arrayZ[-1] + l3)])
        delta_y = 1 / 6 * (k1 + 2 * k2 + 2 * k3 + k4)
        delta_z = 1 / 6 * (l1 + 2 * l2 + 2 * l3 + l4)
        arrayX.append(arrayX[-1] + h)
        arrayY.append(arrayY[-1] + delta_y)
        arrayZ.append(arrayZ[-1] + delta_z)
        if i == v - 1:
                    print("x", i + 1, " = ", arrayX[-1],"  y", i, " = ", arrayY[-1], sep = "")
    print('\n')
    return arrayX, arrayY
```

```python
arrayX1 = [startX]
arrayY1 = [y1]
arrayZ1 = [z1]
arrayX1, arrayY1 = rungeKutta(v=10, h=0.1, arrayX=arrayX1, arrayY=arrayY1, arrayZ=arrayZ1)
```

```python
arrayX2 = [startX]
arrayY2 = [y1]
arrayZ2 = [z1]
arrayX2, arrayY2 = rungeKutta(v=100, h=0.01, arrayX=arrayX2, arrayY=arrayY2, arrayZ=arrayZ2)
```

```python
a = (arrayX1, arrayY1)
b = (arrayX2, arrayY2)
graphic(a=a, b=b)
```
